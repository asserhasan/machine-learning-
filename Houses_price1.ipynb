```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0 , 2.0])
y_train = np.array([300.0 , 500.0])
```

```python
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"the size is {x_train.shape}")
m = x_train.shape[0]
print (f"size of m = {m}")
```

```python
#printing dataset
i = 0
print (f"x^{i} , y^{i} = {x_train[i] , y_train[i]}")
i += 1
print (f"x^{i} , y^{i} = {x_train[i] , y_train[i]}")
```

```python
plt.scatter(x_train , y_train, marker = "x",c = "red")
plt.title("Houses prices model")
plt.ylabel("price (in 1000s of dollars)")
plt.xlabel("size(1000 sqft)")
```

```python
w = 200
b = 100
```

```python
def compute_model_output(x, w, b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b

  return f_wb
```

```python
temp_f_wb = compute_model_output(x_train , w, b)
```

```python
plt.plot(x_train, temp_f_wb, c='b', label="prediction")
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
x_predict = np.array([1.2])
compute_model_output(x_predict,w,b)
```
